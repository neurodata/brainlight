```python
import numpy as np
import brainlit
from brainlit.algorithms.trace_analysis.spline_fxns import curvature, torsion
from brainlit.algorithms.trace_analysis.fit_spline import GeometricGraph
from brainlit.utils import swc
from cloudvolume.exceptions import SkeletonDecodeError
import os

# EXTRACT DATA DIR FROM RELATIVE NOTEBOOK PATH
cwd = os.getcwd()
notebooks_dir = os.path.dirname(cwd)
docs_dir = os.path.dirname(notebooks_dir)
repo_root = os.path.dirname(docs_dir)
data_dir = os.path.join(repo_root, "data")
poster_reproduction_data_dir = os.path.join(data_dir, "poster_reproduction")
segments_dir = os.path.join(poster_reproduction_data_dir, "segments")
```

## Download segments and save them

```python
url_segments = "s3://open-neurodata/brainlit/brain1_segments"

class controlled_read_s3():

    def __enter__(self):
        return 

    def __exit__(self, type, value, traceback):
        return isinstance(value, SkeletonDecodeError)

max_id = 300
mip = 0
ids = []
df_neurons = []
for i in np.arange(0, 300):
    i = int(i)
    with controlled_read_s3():
        df_neuron = swc.read_s3(url_segments, seg_id=i, mip=0)
        df_neurons.to_csv(os.path.join(segments_dir, "{}.csv".format(i)))
        ids.append(i)
```
