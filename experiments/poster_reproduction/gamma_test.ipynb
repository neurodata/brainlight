```python
import matplotlib.pyplot as plt
import scipy.stats
import numpy as np

xx = np.linspace(1, 10000, 1000)
plt.plot(np.log(xx), scipy.stats.gamma.pdf(xx, 325.39752001559077, -1.398343072717216, 0.004294015861780276))
```

```python
def func(x, a, b, c):
    return [c * np.exp(-a*(u-b)) if u > b else 0 for u in x]

xx = np.linspace(-300, 30, 1000)
plt.plot(xx, func(xx, 0.7842297022767486, -136.71040485263688, 137.51256554027782))
```

```python
log_x = np.linspace(0, 10)
log_y = -0.5*log_x - 1.19
plt.plot(log_x, log_y)
```

```python
y = np.exp(log_y)
plt.plot(log_x + 2, y)
```

```python
dist = scipy.stats.powerlaw(0.35458013328375937, loc=-6.0633568457555e-27, scale=17.077328014360464)
x = np.exp(np.arange(0, 10))
print(np.log(dist.pdf(x)))
plt.plot(np.log(x), np.log(dist.pdf(x)))
```

```python
fig = plt.figure(figsize=(12, 8))
axes = fig.subplots(1, 2)
GRAY = "#999999"
TITLE_TYPE_SETTINGS = {"fontname":"Arial", "size":18}
SUP_TITLE_TYPE_SETTINGS = {"fontname": "Arial", "size":22}
plt.rc('font', family="Arial", size=18)
for ax in axes:
    ax.spines['bottom'].set_color(GRAY)
    ax.spines['top'].set_color(GRAY) 
    ax.spines['right'].set_color(GRAY)
    ax.spines['left'].set_color(GRAY)
    ax.tick_params(axis='both', colors=GRAY, labelsize="large")
    ax.set_title("This is a title")
    xx = np.linspace(1, 10)
    ax.plot(xx, xx, color="#e41a1c", lw=2, label="line")
    ax.scatter(xx, xx/2, color="#377eb8", label="data")
    leg = ax.legend()
    leg.get_frame().set_edgecolor(GRAY)
    ax.set_xlabel("x axis")
    ax.set_ylabel("y axis")
fig.suptitle("This is the sup title")
```
